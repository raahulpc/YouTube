**Q13.Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.**

```python
df['dislikes rate']=df['dislikes']/df['view_count']*100
pd.DataFrame(df[['dislikes rate','view_count']]).plot(x='view_count',y='dislikes rate',kind='scatter')
```

```python
sns.heatmap(df.corr(),annot=True,cmap='RdBu')
```

**The scatter plot tells us that as the number of views increase the number of dislikes also increase and therefore we can say that there is a correlation between view count and dislikes.
The heatmap also corroborates the findings of the scatterplot by showing a moderate/high correlation of 0.68 between dislikes and view count.**
